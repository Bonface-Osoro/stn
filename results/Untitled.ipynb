```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("windtexter_results.csv")
df1 = df.groupby(['windtexter','interception_probability'])['iteration'].count()
df1.to_csv("grouped.csv")
df1
```

```python
df1 = pd.read_csv("grouped.csv")
df1.columns = ['windtexter', 'interception_probability','iteration']

long_data = pd.melt(df1,
    id_vars=['windtexter', 'interception_probability'],
    value_vars=['iteration'])

long_data.columns = ['windtexter', 'interception_probability', 'Metric', 'Value']
sns.set(font_scale=1)
plot=sns.barplot(x="windtexter", y='Value', linewidth=1.2, hue='interception_probability',
    palette=sns.color_palette("bright", 9),
    data=long_data)
```
